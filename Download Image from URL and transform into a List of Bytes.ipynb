```python
from IPython.core.display import display, HTML
import requests
import re

from io import BytesIO
from PIL import Image
```

```python
def downloadImage(url):
    res = requests.get(url, stream = True)
    if res.status_code == 200:
        res.raw.decode_content = True
        img = Image.open(res.raw)
        img_barray = BytesIO()
        img.save(img_barray, format=img.format)
        return list(img_barray.getvalue())
    return []
```

```python
def testDownloadAndDisplay():
    url = "https://cdn.myanimelist.net/images/anime/1223/96541.jpg"
    bytesList = downloadImage(url)
    display(len(bytesList))
    display("from list to image again!")
    display(Image.open(BytesIO(bytearray(bytesList))))
    
```

```python
testDownloadAndDisplay()
```
